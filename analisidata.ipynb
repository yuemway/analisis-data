## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
byseason_df = all_df.groupby(by="season").cnt.mean().reset_index()

# Ubah angka season menjadi nama musim
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
byseason_df['season'] = byseason_df['season'].map(season_mapping)


byseason_df.rename(columns={
    "cnt": "total_rentals"
}, inplace=True)

# Membuat grafik bar plot
plt.figure(figsize=(10, 5))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(
    x="total_rentals",
    y="season",
    data=byseason_df.sort_values(by="total_rentals", ascending=False),
    palette=colors
)

plt.title("Total rental sepeda berdasarkan musim", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

### Pertanyaan 2:

```python
byday_df = all_df.groupby(by=["holiday", "workingday"]).cnt.mean().reset_index()

byday_df.rename(columns={
    "cnt": "total_rentals"
}, inplace=True)

byday_df['holiday'] = byday_df['holiday'].replace({0: 'Workingday', 1: 'Holiday'})

# Membuat grafik bar plot
plt.figure(figsize=(10, 5))
colors = ["#72BCD4", "#D3D3D3"]

sns.barplot(
    x="total_rentals",
    y="holiday",
    hue="workingday",
    data=byday_df.sort_values(by="total_rentals", ascending=False),
    palette=colors
)

plt.title("Rata-rata jumlah rental sepeda berdasarkan workingday dan holiday", loc="center", fontsize=15)
plt.ylabel("Holiday Status")
plt.xlabel("Rata-rata jumlah rental")
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

**Insight:**
- Rental sepeda paling banyak digunakan saat fall >> summer >> winter >> spring
- Sedangkan jika berdasarkan holiday dan workingday, rental sepeda paling banyak digunakan saat holiday di workingday >> workingday >> holiday

Ini menunjukkan bahwa meskipun hari libur, jika bertepatan dengan hari kerja, permintaan sepeda tetap tinggi, sementara pada hari libur biasa, penggunaannya cenderung lebih rendah.
